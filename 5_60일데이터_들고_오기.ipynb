### 작동

```python
# 닭찌찌
post_date1 = []
review1 = []  # 데이터를 담을 리스트
url1 = 'https://www.kurly.com/shop/goods/goods_review_list.php?goodsno=5059674&page='
a = soup.find_all('a')[-1]
last_page = a['href'].split('=')[-1]
last_page = int(last_page)
print(type(last_page))
i = 0
while True:    # 무한 루프
    url = url1 + str(i)
    driver.get(url)
    i += 1          # i를 1씩 증가시킴

    html_str = driver.page_source
    soup = BeautifulSoup(html_str)
    target_list = soup.select('div.inner_review')
    number = soup.select('tr')


    try:    # 오류시 넘기기 위해
        for j in range(0,7):
            num = number[j+1].get_text().split('\n\n')
            if num[1] != '공지' and num[1] != 'BEST'  :     # 공지글과 BEST글은 스킵
                date = number[j+1].text.split('\n')
                b=date[-5]
                c=date[-5].split('-')
                timeget2 = datetime.date(int(c[0]),int(c[1]),int(c[2]))
                print(twomonths,timeget2 )
                if twomonths <= timeget2:
                    post_date1.append(b)
                    a = target_list[j].get_text().split('\n\n\n')
                    review1.append(a[1].replace('\n',''))  # 리뷰에 담기
                    if twomonths >= timeget2:  # 이틀전 날짜가 글 게시 날짜와 같은때
                        break
                else:
                    continue

    except IndexError :
        print('잘못된 인덱스입니다')  # 오류시 나오는 글

    if twomonths >= timeget2: # 이틀전 날짜가 글 게시 날짜와 같을 때
        break
```

```python
post_date1
```

```python
data_dict1 = {
    'sentence' : [],
    'date_posted' : [],
    'Market_pid' : []
}

data_dict1['sentence'] = review1
data_dict1['date_posted'] = post_date1
data_dict1['Market_pid'] = '5059674'
df1 = pd.DataFrame(data_dict1)
df1
```

```python
df1.isna().sum()
```

```python
# 저장
df1.to_csv('./data/두달닭찌찌.csv', encoding = 'utf-8-sig', index = False)
```
